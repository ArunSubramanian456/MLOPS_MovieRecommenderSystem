```python
import pandas as pd
import numpy as np
import ast

import matplotlib.pyplot as ply
import seaborn as sns
```

```python
credits = pd.read_csv('./data/credits.csv')
credits.head()
```

```python
meta = pd.read_csv('./data/movies_metadata.csv')
meta.head()
```

```python
credits.info()
```

```python
meta.info()
```

```python
# Convert release date to pandas datetime type
meta['release_date'] = pd.to_datetime(meta.release_date, errors='coerce')
```

```python
# Create a year column
meta['year'] = meta['release_date'].dt.year
```

```python
# Check the number of movies by year
meta['year'].value_counts().sort_index()
```

```python
# Get data only for 2017 movies; we already have data until 2016 from Part 1 data processing

new_meta = meta.loc[meta.year ==2017, ['genres','id','title','year']]
new_meta.head()
```

```python
# Convert id to integer
new_meta['id'] = new_meta.id.astype(int)
```

```python
# Merge credits data to new metadata

data = pd.merge(new_meta, credits, on='id')
data.head()
```

```python
# Convert genres, cast and crew columns into a list

data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))  #the "literal_eval" helps to convert the string into a list
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))

data.head()
```

```python
# Now we will have to take out the individual genres like adventure, action, sci-fi, etc. using the following function


def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction': #am only renaming the "Science Fiction" to "Sci-Fi" to make it the name shorter. Apart from that all other names remain the same.
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen)) #then we will join them together and return the valuse
```

```python
# lets apply the function to genres column

data['genres_list'] = data['genres'].map(lambda x : make_genresList(x))
data.head()
```

```python
# Function to clean crew column

def get_actor(x, n=0):
    casts = []
    for i in x:
            casts.append(i.get('name'))
    if casts == [] or len(casts)<=n :
        return np.NaN
    else:
        return (casts[n]) #then we will join them together and return the valuse
```

```python
# Add Actor 1 name column
data['actor_1_name'] = data['cast'].map(lambda x : get_actor(x,0))
data.head()
```

```python
# Add Actor 2 name column
data['actor_2_name'] = data['cast'].map(lambda x : get_actor(x,1))
data.head()
```

```python
# Add Actor 3 name column
data['actor_3_name'] = data['cast'].map(lambda x : get_actor(x,2))
data.head()
```

```python
# Function to get Director name

def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
# Add Director name column

data['director_name'] = data['crew'].map(lambda x: get_directors(x))
data.head()
```

```python
# Selecting only the prepared data

cols = ['director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'genres_list','title']

movie = data.loc[:, cols]
movie.head()
```

```python
# Check for missing values
movie.isnull().sum()
```

```python
# drop missing values

movie = movie.dropna(how='any')
```

```python
# Check if missing value records are dropped
movie.isnull().sum()
```

```python
# Rename columns
movie = movie.rename(columns = {'genres_list' : 'genres', 'title' : 'movie_title'})
movie.head()
```

```python
# Convert the movie title column to lower case

movie['movie_title'] = movie['movie_title'].str.lower()
```

```python
#we will be using this information later on in the tfidvectorizor

movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']
movie.head()
```

```python
# Get data with movies until 2016

movie2016 = pd.read_csv('./data1.csv')
```

```python
# Create combination column for movie2016 data
movie2016['comb'] = movie2016['actor_1_name'] + ' ' + movie2016['actor_2_name'] + ' '+ movie2016['actor_3_name'] + ' '+ movie2016['director_name'] +' ' + movie2016['genres']
movie2016.head()
```

```python
# Putting all datasets together

new = pd.concat([movie2016, movie], axis=0)
new.info()
```

```python
# Delete duplicate records
new.drop_duplicates(subset = "movie_title", keep = 'last', inplace=True)
new.info()
```
